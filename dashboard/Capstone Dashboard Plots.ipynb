```python
grouped_by_year = tb_df.groupby(["Year"]).sum()
```

```python
deathratesbyyear= grouped_by_year.drop(['Country/Territory','Code'], axis=1)
deathratesbyyear.sample(5)
```

```python
top_causes = deathratesbyyear.sum().sort_values(ascending=False).head(10)

fig = px.bar(top_causes, x=top_causes.index, y=top_causes.values, labels={'x': 'Causes of Death', 'y': 'Number of Deaths'})
fig.update_layout(title='Top 10 Causes of Death (1990-2019)', xaxis_title='Causes of Death', yaxis_title='Number of Deaths')

fig.show()
```

```python
fig = px.line(data_frame=deathratesbyyear, x=deathratesbyyear.index, y="Tuberculosis")
fig.show()
```

```python
highest_tuberculosis_deaths = deathratesbyyear['Tuberculosis'].idxmax()

highest_tuberculosis_deaths_data = deathratesbyyear.loc[highest_tuberculosis_deaths]

# Print or display the result
print("Year with the highest Tuberculosis deaths:", highest_tuberculosis_deaths)
print("\nDetails:")
print(highest_tuberculosis_deaths_data)
```

```python
fig = px.line(tb_df, x='Year', y='Tuberculosis', color='Country/Territory', title='Tuberculosis Deaths Over Time')
fig.show()
```

```python
top_countries_tuberculosis = tb_df.groupby('Country/Territory')['Tuberculosis'].sum().sort_values(ascending=False).head(10)
fig = px.bar(top_countries_tuberculosis, x=top_countries_tuberculosis.index, y=top_countries_tuberculosis.values, title='Top 10 Countries with Tuberculosis Deaths')
fig.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

tb_deaths_by_country_year = tb_df.pivot_table(values='Tuberculosis', index='Country/Territory', columns='Year', aggfunc='sum')
plt.figure(figsize=(15, 12))
sns.heatmap(tb_deaths_by_country_year, cmap='YlGnBu', linewidths=.5)
plt.title('Tuberculosis Deaths by Country and Year')
plt.show()
```
